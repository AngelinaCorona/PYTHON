```python
import pandas as pd
students = pd.read_csv("students.csv")
print(students.info())


# Mostrar el typo de dato
print(students.dtypes)
```

```python
# Estadisticas basicas de columnas numericas
print(students.describe())
```

```python
# Descripcion de columnas no numericas
print(students.describe(include = 'objetc'))
```

```python
# Seleccionar columnas y filas

# Seleccionar por columna:
print(students["math score"])

# Obtener la cantidad de alumnos por genero
print(students['gender'].value_counts())

# Seleccionar varias columnas:
print(students[["math score", "gender"]])
```

```python
# Filtrar datos con condicion

# Alumnos que obtuvieron 100 en matematicas
Cal_100 = students[students['math score'] == 100]
print(Cal_100)
```

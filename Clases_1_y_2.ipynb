## Funciones en python

```python
"""
Se definen con la palabra def
se reciben parametros
Devuelven valor "return"

Sintaxis

def nombre_de_funcion (param1, param2, etc...):
  Bloque de codigo
  return valor
  print()
"""

# Ejemplo 1
# Funcin para sumar dos numeros

def suma(a,b):
  c= a+b
  return c

print(suma(2,3))

def suma_2(a,b):
  return a+b

print(suma_2(2,3))


# Definir valores predeterminados

def suma_3(a,b=1):
  return a+b

print (suma_3(1))
print (suma_3(1,2))


```

```python
# Ejemplo de multiplicar un numero x2

def duplicar(a):
  x = a*2
  return x

print(duplicar(4))

# Funciones con multiples resultados

def operaciones(a,b):
  suma = a+b
  resta = a-b
  return suma,resta

resultado = operaciones(3,2)
print(resultado)
```

```python
# Funciones con cadenas de texto

#upper

def mayusculas(texto):
  return texto.upper()

print (mayusculas("hola mundo"))

#lower

def minusculas(texto):
  return texto.lower()

print (minusculas("HOLA MUNDO"))

# funiones usando listas
# Lista y quiero que mi funcion me devuelva el numero mas grande

def maximo(lista):
  return max(lista)

num = [1,2,3,4]

print(maximo(num))
print(maximo([1,2,3,4]))

# Funciones con diccionarios
# Ejemplo

persona = {
    "Nombre":"Carlos",
    "Edad":15
}

print(persona)
print(persona.get("Nombre"))

# Metodo .get
# Este metodo devuelve el valor de la clave que especifiques
# Opcional: asignar un valor en caso de que tu clave no exista

#Ejemplo 1

def saludo(diccionario, clave):
  x = diccionario.get(clave)
  return f"Hola {x}"

print (saludo(persona, "Nombre"))

#Ejemplo 2 con opcion de que no haya dato

def saludo(diccionario, clave):
  x = diccionario.get(clave, "No existe llave")
  return f"Hola {x}"

print (saludo(persona, "Ciudad"))


# Ejemplo 3

personas = {
    "Nombre":["Carlos","Juan","Maria"],
    "Edad":[30,24,16]
}

def saludo(diccionario,clave,orden):
  x = diccionario.get(clave)[orden]
  return f"Hola {x}"

print(saludo(personas,"Nombre",0))
print(saludo(personas,"Nombre",2))

```

```python
"""
Ejercicio Funciones

Escribe una funcion que reciba una lista de numeros
y regrese el numero mayor max() y el menor min()

x = [11,25,30]

"""

def resultado(lista):
  return max(lista), min(lista)
x = [11,25,30]

print(resultado(x))

```

```python
# Ejemplo de funciones con un ciclo for

"""
Crear una tabla de multiplicar
del 1 al 10
"""

def tabla_mult(num):
  for i in range (0,11):
    print(f"{num} x {i} = {num*i}")

num = int(input("Muestrame la tabla del: "))

tabla_mult(num)

```

```python
# Estadisticas basicas

arr = np.array([5,10,15,20,25])

print("Promedio", np.mean(arr))
print("Mediana", np.median(arr))
print("Desviacion estandar", np.std(arr))
print("Suma",np.sum(arr))
print("Maximo", np.max(arr))
print("Minimo", np.min(arr))
```

## Visualizacion de datos

### Matplotlib

```python
import matplotlib.pyplot as plt

# Visualizar el promedio por genero (calificacion de matematicas)

promedio_por_genero = students.groupby("gender")["math score"].mean()
print(promedio_por_genero)

# Graficar
promedio_por_genero.plot(
    kind = 'bar',
    figsize = (6,4),
    color = ['blue','red']
)

plt.title("Promedio de matematicas por genero")
plt.xlabel("Genero")
plt.ylabel("Promedio")
plt.show()
```

```python
# graficos de dispersion

students.plot(
    kind = 'scatter',
    x = 'reading score',
    y = 'writing score',
    color = 'green',
    figsize = (10,5)
)

plt.title("Relacion entre reading score y writing")
plt.xlabel("Reading")
plt.ylabel("Wiriting")
plt.show()
```

```python
# Histograma
"""
Sirven para saber la distribucion de los datos
bins es el numero de barras que va a tener mi histograma
y es como el ancho de la barra
"""

students["math score"].plot(
    kind = 'hist',
    bins = 100,
    color = 'purple',
    edgecolor = 'black',
    figsize = (10,5)
)
plt.title("Distribucion de cal en matematicas")
plt.xlabel("Puntaje")
plt.ylabel("Frecuencia")
plt.show()
```

### Seaborn

```python
import seaborn as sns

# Ejemplo 1
sns.scatterplot(
    x = "reading score",
    y = "writing score",
    data = students
)

plt.title("Relacion entre writing y reading")
plt.show()

# Ejemplo 2 (que cada punto me dig auna variable categorica)
sns.scatterplot(
    x = "reading score",
    y = "writing score",
    data = students,
    hue = "gender"
)

plt.title("Relacion entre writing y reading")
plt.show()
```

```python
# Histograma usando seaborn

sns.histplot(students["math score"],
             bins = 10,
             kde = True,
             color = 'purple'
             )
plt.title("Distribucion de cal de matematicas")
```

```python
# Boxplot

sns.boxplot(
    x = "gender",
    y = "math score",
    data = students,
    palette = "Set2",
    hue = "test preparation course"
)

plt.title("Diagrama de caja - Distribucion de cal por genero")
plt.show()
```
